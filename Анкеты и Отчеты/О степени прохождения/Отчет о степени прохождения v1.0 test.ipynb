```python
def make_table(tab_in, key):
    nice, case, know, test, data, excel_table, excel_table_new = tab_in
    result = nice[nice['referer'] == key][['name', 'name_2', 'name_3', 'ma_email', 'phone', 'univer', 'univer_2']]
    #result = result.sort_values('name', ascending=False).drop_duplicates('ma_email').sort_index().reset_index(drop=True)
    result['Анкета'] = result['ma_email'].apply(lambda x: find_if(x, nice, 'ma_email', 0, str))
    result['Кейсы'] = result['ma_email'].apply(lambda x: find_if(x, case, 'case1', 10, str))
    result['Материал'] = result['ma_email'].apply(lambda x: find_if(x, know, 'vybormateriala', 10, str))
    #result['Тест'] = result['ma_email'].apply(lambda x: find_if(x, test, 'score', 0, int))
    result['Новелла'] = result['ma_email'].apply(lambda x: find_if(x, data, 'history_new', 20, str))
    result['Процент'] = result[['Кейсы', 'Материал', 'Анкета', 'Новелла']].apply(lambda x: sum(x), axis=1) * 25

    c_1 = 0
    df_1 = excel_table_new[excel_table_new['referer'] == key]
    for _d in df_1.to_dict(orient="records"):
        if _d['email'] not in result['ma_email'].values:
            result.loc[len(result)] = [_d['name'], _d['name_2'], '', _d['email'], _d['phone'], '', _d['univer'], 0, 0, 0, 0, 0]
            c_1 += 1  
    print('Новых нулевых: {}'.format(c_1))
    
    c_2 = 0
    df_2 = excel_table.parse(values['-KEY-'])
    for name, email in df_2.values:
        if email not in result['ma_email'].values:
            result.loc[len(result)] = [name, '', '', email, '', '', '', 0, 0, 0, 0, 0]
            c_2 += 1  
    print('Старых нулевых: {}'.format(c_2))

    result = result[~result['ma_email'].isin(['test@technospark.ru', 'test@edu.technospark.ru'])]
    
    return result
```

```python
def find_if(email, table, column, level, type_):
    if type_ == str:
        return (1 if len(table[table['ma_email'].apply(str.lower) == email.lower()][column]) and len(max(table[table['ma_email'].apply(str.lower) == email.lower()][column])) > level else 0)
    else:
        return (1 if len(table[table['ma_email'].apply(str.lower) == email.lower()][column]) and (max(table[table['ma_email'].apply(str.lower) == email.lower()][column])) != level else 0)
```
